Tải bộ dữ liệu

```python
import pandas as pd
import os

# Đường dẫn đến dataset cũ
dataset_path_old = "/kaggle/input/datasets/atifaliak/youtube-comments-dataset"
dataset_path_new = "/kaggle/input/datasets/kienlytran/youtube-comment-3k6-neutral-3k6-negative"

# Kiểm tra các file
print("Các file trong dataset cũ:")
for file in os.listdir(dataset_path_old):
    print(f"  - {file}")

print("\nCác file trong dataset mới:")
for file in os.listdir(dataset_path_new):
    print(f"  - {file}")
```

```python
df_old = pd.read_csv(f"{dataset_path_old}/YoutubeCommentsDataSet.csv")
df_new = pd.read_csv(f"{dataset_path_new}/neutral_3600_negative_all.csv")
print("\n=== DỮ LIỆU CŨ ===")
print(f"Shape: {df_old.shape}")
print("Phân bố nhãn:")
print(df_old['Sentiment'].value_counts())

print("\n=== DỮ LIỆU MỚI ===")
print(f"Shape: {df_new.shape}")
print("Phân bố nhãn:")
print(df_new['Sentiment'].value_counts())
```

```python
# Gộp dữ liệu
df = pd.concat([df_old, df_new], ignore_index=True)

print("\n=== DỮ LIỆU SAU KHI GỘP ===")
print(f"Shape: {df.shape}")
print("Phân bố nhãn:")
print(df['Sentiment'].value_counts())
print("\nTỷ lệ (%):")
print(df['Sentiment'].value_counts(normalize=True).round(4) * 100)
```

```python
# Tiền xử lý dữ liệu
df = df.dropna(subset=['Comment'])

clean_df = pd.DataFrame({
    'Sentence': df['Comment'],
    'Emotion': df['Sentiment']
})

print(f"\nShape sau khi làm sạch: {clean_df.shape}")
```

```python

# Chia dữ liệu train/validation/test
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    clean_df, 
    test_size=0.3, 
    random_state=42,
    stratify=clean_df['Emotion']
)

valid_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    random_state=42,
    stratify=temp_df['Emotion']
)

print(f"\nTrain: {len(train_df)}")
print(f"Validation: {len(valid_df)}")
print(f"Test: {len(test_df)}")

```

Thống kê cơ bản bộ dữ liệu

```python
# Thống kê phân bố
print("\n=== PHÂN BỐ NHÃN TOÀN BỘ ===")
print(clean_df['Emotion'].value_counts())

```

```python
print("\n=== PHÂN BỐ NHÃN TRAIN ===")
print(train_df['Emotion'].value_counts())

```

```python
print("\n=== PHÂN BỐ NHÃN VALIDATION ===")
print(valid_df['Emotion'].value_counts())

```

```python
print("\n=== PHÂN BỐ NHÃN TEST ===")
print(test_df['Emotion'].value_counts())
```

```python
# Lưu dữ liệu
output_path = '/kaggle/working/youtube_comments_prepared'
os.makedirs(output_path, exist_ok=True)

train_df.to_csv(f"{output_path}/youtube_train.csv", index=False, encoding="utf-8-sig")
valid_df.to_csv(f"{output_path}/youtube_valid.csv", index=False, encoding="utf-8-sig")
test_df.to_csv(f"{output_path}/youtube_test.csv", index=False, encoding="utf-8-sig")
clean_df.to_csv(f"{output_path}/youtube_merged.csv", index=False, encoding="utf-8-sig")
train_valid_df = pd.concat([train_df, valid_df], ignore_index=True)
train_valid_df.to_csv(f"{output_path}/youtube_merged_train_valid.csv", index=False, encoding="utf-8-sig")

print(f"\nĐã lưu vào: {output_path}")
print("Các file:")
for f in os.listdir(output_path):
    print(f"  - {f}")

```

```python
# Mẫu dữ liệu
print("\n=== MẪU DỮ LIỆU ===")
for sentiment in ['positive', 'negative', 'neutral']:
    samples = clean_df[clean_df['Emotion'] == sentiment]['Sentence'].head(2)
    print(f"\n{sentiment.upper()}:")
    for sample in samples:
        print(f"  - {sample[:100]}{'...' if len(sample) > 100 else ''}")
```
